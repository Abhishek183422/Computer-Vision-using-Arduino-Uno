```python
import serial
import cv2
import numpy as np
from tensorflow import keras
```

```python
model = keras.models.load_model('/Users/draculalemon/Documents/arduino_sign.h5')
```

```python
ser=serial.Serial('/dev/cu.usbmodem1101',9600)
print('Connected to Port')
```

```python

vide=cv2.VideoCapture(0)
while True:
    ret,frame=vide.read()
    imgr1=cv2.resize(frame,(200,200))
    imgrgb1=cv2.cvtColor(imgr1,cv2.COLOR_BGR2RGB)
    final_format1=np.array([imgrgb1]).astype('float64')/255.0  
    predd=model.predict(final_format1)
    index=np.argmax(predd[0])
    new_proba=str(round(np.max(predd[0]*100),2))
    if new_proba >= "99.00":
        label_name=['go','stop']
        labels=label_name[index]
        cv2.putText(frame,f'This is a {labels} {new_proba}%',(50,50),1,3,(100,20,255),4)
        #arduino code 
        if labels=='go':
            my_cmd=labels
            my_cmd=my_cmd+'\r'
            ser.write(my_cmd.encode())
        elif labels=='stop':
            my_cmd=labels
            my_cmd=my_cmd+'\r'
            ser.write(my_cmd.encode())
                        
    else:
        cv2.putText(frame,'Nothing detected',(50,50),1,3,(100,20,255),4)

    #gettig the class name
    #write predicted flower name in the image
   
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
vide.release()
cv2.destroyAllWindows()
```
